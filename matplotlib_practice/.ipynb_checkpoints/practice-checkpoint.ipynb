```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
# set the x-axis to a list of strings for each month
x_axis=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

#set the y-axis to a list of flots as the total fare in US dollars accumulated each month
y_axis=[10.02,23.24,39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# another way to create the plot with ax.plt
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create yet another plot using ax.plt
ax=plt.axes()
ax.plot(x_axis,y_axis)
```

```python
#create the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston', marker='*',color='blue',linewidth=2)
#Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
#Set the y limit between 0 and 45
plt.ylim(0,45)
#Create a title
plt.title('PyBer Fare by Month')
# Add the legend
plt.legend()
```

```python
# Create the plot with ax.plt
fig, ax = plt.subplots()
ax.grid()
ax.plot(x_axis, y_axis,label='Boston',color='green',marker="D")
ax.legend()
ax.set_ylabel('Fares($)')
ax.set_xlabel('Month')
```
